```python
# full disclosure: I gave the outerhtml to chat GPT to figure out the correct tags as we did in the end of the class
# I needed it after the Le Monde struggles


import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get("https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations/")
doc = BeautifulSoup(response.text)
```

```python
table = doc.select_one('table.table')     
items = table.select('tbody tr')
```

```python
rows = []

for item in items:
    row = {}

    # Company Name
    try:
        company_tag = item.select_one('td:nth-of-type(1) a')
        row['company'] = company_tag.text.strip() if company_tag else item.select_one('td:nth-of-type(1)').text.strip()
    except:
        row['company'] = 'no company'

    # Company URL
    try:
        row['url'] = item.select_one('td:nth-of-type(1) a').get('href')
    except:
        row['url'] = None

    # Address
    try:
        row['address'] = item.select_one('td:nth-of-type(2)').get_text(separator=' ', strip=True)
    except:
        row['address'] = 'no address'

    # Telephone
    try:
        row['telephone'] = item.select_one('td:nth-of-type(3)').get_text(separator=' ', strip=True)
    except:
        row['telephone'] = 'no telephone'

    # Hours
    try:
        row['hours'] = item.select_one('td:nth-of-type(4)').get_text(separator=' ', strip=True)
    except:
        row['hours'] = 'no hours'

    rows.append(row)
```

```python
len(rows)
```

```python
df = pd.json_normalize(rows)
```

```python
df.to_csv('driver_licence.csv')
```
